```python
import csv

# Merge sort implementation
def merge_sort(arr, compare_func):
    if len(arr) <= 1:
        return arr

    # Split the array into two halves
    middle = len(arr) // 2
    left_half = arr[:middle]
    right_half = arr[middle:]

    # Recursively sort both halves
    left_half = merge_sort(left_half, compare_func)
    right_half = merge_sort(right_half, compare_func)

    # Merge the sorted halves
    sorted_arr = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left_half) and right_idx < len(right_half):
        if compare_func(left_half[left_idx], right_half[right_idx]):
            sorted_arr.append(left_half[left_idx])
            left_idx += 1
        else:
            sorted_arr.append(right_half[right_idx])
            right_idx += 1

    # Append any remaining elements from both halves
    sorted_arr.extend(left_half[left_idx:])
    sorted_arr.extend(right_half[right_idx:])

    return sorted_arr

# Custom comparison function based on the number of rides
def compare_by_rides(traveler1, traveler2):
    return traveler1['Number_of_Rides'] > traveler2['Number_of_Rides']

# Load data from CSV files
north_zone_data = []
south_zone_data = []

with open("/content/drive/MyDrive/north_zone_data.csv", mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        north_zone_data.append(row)

with open("/content/drive/MyDrive/south_zone_data.csv", mode='r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        south_zone_data.append(row)

all_travelers = north_zone_data + south_zone_data

# Sort the list of travelers using merge sort
sorted_travelers1 = merge_sort(north_zone_data, compare_by_rides)
sorted_travelers2 = merge_sort(south_zone_data, compare_by_rides)
sorted_travelers = merge_sort(all_travelers, compare_by_rides)

# Print the sorted list of travelers
print("Frequent travellers in North zone\n")
for traveler in sorted_travelers1[:5]:
    print(f"Name: {traveler['Name']}, Number of Rides: {traveler['Number_of_Rides']}")

print("\nFrequent travellers in South Zone\n")
for traveler in sorted_travelers2[:5]:
    print(f"Name: {traveler['Name']}, Number of Rides: {traveler['Number_of_Rides']}")


print("\nFrequent travellers in Both Zones\n")
for traveler in sorted_travelers[:5]:
    print(f"Name: {traveler['Name']}, Number of Rides: {traveler['Number_of_Rides']}")
```
